```python
anime_with_synopsis = pd.read_csv("data/anime_with_synopsis.csv",)# index_col='MAL_ID')
```

```python
import requests, re, time

for i in anime_with_synopsis.loc[anime_with_synopsis.Score == 'Unknown', 'Name'][:100]: # first 100
    getIndex = anime_with_synopsis.loc[(anime_with_synopsis.Score == 'Unknown') & ((anime_with_synopsis.Name == i))].index[0]
    time.sleep(2)
    getRequests = requests.get(f"https://myanimelist.net/anime/{getIndex}")
    getScore = re.findall(r"<div *class=.score-label.score.[0-9]+.*?>(.*?)</div>", getRequests.text)
    if len(getScore) == 0:
        getScore.append(0.0)
    getScore = float(getScore[0])
    getUnknownCols = anime_with_synopsis.loc[getIndex]['Score']
    #print(getUnknownCols, getScore, getIndex)
    #getUnknownCols = getUnknownCols.replace("Unknown", str(getScore))
    anime_with_synopsis.loc[getIndex]['Score'] = anime_with_synopsis.loc[getIndex]['Score'].replace("Unknown", str(getScore))
    
```

```python
anime_with_synopsis.to_csv("anime_with_synopsis_with_score_added_batch1.csv")
```
